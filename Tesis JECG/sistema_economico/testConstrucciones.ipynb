```python
import numpy as np

# ajuste de los individuos

# funcion objetivo
def ajusteEntrada(sujeto):
    demandaReal = 35.39435274190398
    r = sujeto[0]
    p1 = sujeto[1]
    p2 = sujeto[2]
    demandaCalculada = (demandaReal * p1) / (r - 2 + 2*p2 + p1)
    ajuste = demandaCalculada - demandaReal
    return abs(ajuste)


# metodo de seleccion de padres
def seleccionar_padres(poblacion, ajuste, numProgenitores):
    progenitores = np.empty((numProgenitores, poblacion.shape[1]))
    # considerar el numero de desplazamientos
    asacendencia = 0
    #for asacendencia in range(numProgenitores):
    while len(poblacion) > 1:
        #print(ajuste)
        renglonCandidato = np.where(ajuste == np.min(ajuste))[0][0]
        progenitores[asacendencia, :] = poblacion[renglonCandidato, :]
        ajuste = np.delete(ajuste, renglonCandidato, 0)
        poblacion = np.delete(poblacion, renglonCandidato, 0)
    return progenitores

# generacion de hijos
def herencia(padres, descendencia):
    hijos = np.empty(descendencia)
    padreIndex1 = 0
    #for numIndividuo in range(int(len(padres)/2)):
    numIndividuo = 0
    while True:
        #print(padreIndex1, numIndividuo)
        padreIndex2 = padreIndex1 + 1
        if padreIndex1 > len(padres)-1:
            break
        ascendente1 = padres[padreIndex1]
        ascendente2 = padres[padreIndex2]
        #print(ascendente1, ascendente2)
        # se necesitan 3 genes, de dos con tres genes:
        hijos[numIndividuo, 0] = min(ascendente1[0], ascendente2[0])
        hijos[numIndividuo, 1] = min(ascendente1[1], ascendente2[1])
        hijos[numIndividuo, 2] = min(ascendente1[2], ascendente2[2])
        numIndividuo += 1
        padreIndex1 += 2

    return hijos




def determinacionParametros(poblacionBase=36):
    # condiciones iniciales de la poblacion
    anterior = None
    gen = 1
    numGenesAlpBet = 2
    numGenesRate = 1
    comparador1 = np.array([0.05, 0.001, 0.001])
    comparador2 = np.array([0.01, 0.001, 0.001])
    elementosAphjBet = (poblacionBase, numGenesAlpBet)
    elementoRate = (poblacionBase, numGenesRate)
    # poblacion inicial
    genesAlphaBetha = np.random.uniform(0, 0.5, size=elementosAphjBet)
    genesRate = np.random.uniform(low=0, high=0.2, size=elementoRate)
    poblacion = np.concatenate((genesRate, genesAlphaBetha), axis=1)
    # n-sima generacion
    while True:
        numAntecesores = int(len(poblacion)/2)
        resultadosPoblacionBase = np.apply_along_axis(ajusteEntrada, -1, poblacion)
        # el numero de progenitores debe ser divisible entre 2
        # seleccionados por ajuste
        padres = seleccionar_padres(poblacion, resultadosPoblacionBase, numAntecesores)
        # retorno de hijos con base en mejores genes
        detalleDescendencia = (int(numAntecesores/2), 3)  # numero de hijos
        hijos = herencia(padres, detalleDescendencia)
        renglonStop = np.where(resultadosPoblacionBase == np.min(resultadosPoblacionBase))[0][0]
        renglonStop = poblacion[renglonStop, :]
        salida = np.logical_and(renglonStop < comparador1[0],
                        renglonStop < comparador1[1],
                        renglonStop < comparador1[2])
        if True in salida:
            break
        else:
            poblacion = hijos
            anterior = renglonStop
        gen+=1


    confirmacionTasa = np.logical_and(renglonStop < comparador2[0],
                                        renglonStop < comparador2[1],
                                        renglonStop < comparador2[2])
    if True in confirmacionTasa:
        parametros = anterior
    else:
        parametros = renglonStop
    return (parametros, anterior, renglonStop, gen)
```

```python
parametros = (None,None,None,None)

while parametros[0] is None and parametros[1] is None:
    parametros = determinacionParametros()
    print(parametros)

paramTaylor = list(parametros[0])
r = paramTaylor[0]* 100
print(r)
```
